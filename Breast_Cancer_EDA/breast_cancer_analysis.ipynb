## 1. Data Ingestion & Elite Configuration
Loading the global oncology repository and injecting the **Saga/Elite** visual rendering engine for display integrity (Kaggle Standard).

## 2. Infrastructure Efficacy Mapping: The MIR Metric
### 2.1 Mortality-to-Incidence Ratio (MIR) Synthesis
We engineer the **MIR** as a high-fidelity proxy for healthcare system operational capacity. An MIR convergence toward zero signifies optimal screening-to-treatment pipelines.

```python
country_data['MIR'] = country_data['Deaths_2022'] / country_data['New_Cases_2022']
mir_extremes = country_data.sort_values('MIR')
target_display = pd.concat([mir_extremes.head(10), mir_extremes.tail(10)])

plt.figure(figsize=(14, 8))
sns.barplot(data=target_display, x='MIR', y='Country', palette="flare")
plt.title("Healthcare System Efficacy Proxy: Mortality-to-Incidence Ratio", fontweight='bold')
plt.xlabel("MIR (Deaths / Incidence Velocity)")
plt.axvline(0.40, color=VIBRANT_GOLD, linestyle='--', alpha=0.5, label='High-Risk Threshold')
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.2)
plt.show()
```

## 3. The Oncology Efficiency Frontier
### 3.1 Mapping Global Treatment Effectiveness
By plotting the **Incidence Rate** vs. the **Mortality Rate**, we visualize the 'Efficiency Frontier'. Nations falling significantly below the linear trend-line exhibit superior oncological outcomes relative to their cancer burden.

```python
plt.figure(figsize=(12, 8))
sns.regplot(data=country_data, x='Incidence_Rate_Per_100K', y='Mortality_Rate_Per_100K', 
            scatter_kws={'s': 100, 'alpha': 0.6, 'color': VIBRANT_CYAN}, 
            line_kws={'color': VIBRANT_PINK, 'lw': 2, 'ls': '--'})

# Labeling Alpha Performers (High burden, low mortality)
alpha_performers = country_data[country_data['MIR'] < 0.12].head(5)
for i, row in alpha_performers.iterrows():
    plt.text(row['Incidence_Rate_Per_100K']+2, row['Mortality_Rate_Per_100K'], row['Country'], color=VIBRANT_GREEN, fontweight='bold')

plt.title("Oncology Efficiency Frontier: Burden vs. Mortality Capture", fontweight='bold')
plt.xlabel("Incidence Rate (per 100K)")
plt.ylabel("Mortality Rate (per 100K)")
plt.grid(True, alpha=0.15)
plt.show()
```

## 4. Regional Survival Stratification & Diagnostic Yield
### 4.1 Continent-Level Performance Analysis
We aggregate the **Diagnostic Yield** (percentage of Stage I/II detection) across continents to measure the disparity in healthcare coverage.

```python
regional_stats = country_data.groupby('Continent').agg({
    'Stage_I_II_Pct': 'mean',
    'Five_Year_Survival_Pct': 'mean',
    'Mammography_Coverage_Pct': 'mean',
    'Treatment_Access_Score': 'mean'
}).sort_values('Five_Year_Survival_Pct', ascending=False)

plt.figure(figsize=(14, 7))
ax = regional_stats[['Stage_I_II_Pct', 'Five_Year_Survival_Pct']].plot(kind='bar', figsize=(12, 6), colormap='cool', edgecolor='white')
plt.title("Regional Survival & Diagnostic Yield Disparity", fontweight='bold')
plt.ylabel("Percentage %")
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

display(regional_stats)
```

## 5. Economic Stratification & The 'Survival Cliff'
### 5.1 Ten-Year Survival Decay Modeling
Evaluating the longitudinal decay of survival probabilities across socioeconomic income regions for Stage II pathologies.

```python
survival_melted = survival_stats.melt(id_vars=['Stage', 'Income_Region'], 
                                      value_vars=['One_Year_Survival_Pct', 'Five_Year_Survival_Pct', 'Ten_Year_Survival_Pct'],
                                      var_name='Period', value_name='Survival_Pct')

period_mapping = {'One_Year_Survival_Pct': 1, 'Five_Year_Survival_Pct': 5, 'Ten_Year_Survival_Pct': 10}
survival_melted['Timeline_Years'] = survival_melted['Period'].map(period_mapping)

plt.figure(figsize=(14, 7))
sns.lineplot(data=survival_melted[survival_melted['Stage'] == 'Stage II'], 
             x='Timeline_Years', y='Survival_Pct', hue='Income_Region', 
             marker='o', linewidth=3.5, palette="spring")

plt.title("Longitudinal Survival Decay (Stage II) by Income Stratum", fontweight='bold')
plt.ylabel("Survival Probability (%)")
plt.xlabel("Years Post-Diagnosis")
plt.ylim(0, 105)
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

## 6. Oncological Risk Scoring Engine
### 6.1 Interactive Risk Prioritization Module
Based on the **Population Attributable Fraction (PAF)**, we develop an interactive scoring simulator to quantify the relative impact of lifestyle factors.

```python
def oncological_risk_engine(factors_list):
    """
    Simulates cumulative modifiable risk based on weighted PAF contribution.
    """
    risk_sum = risk_factors[risk_factors['Risk_Factor'].isin(factors_list)]['Population_Attributable_Fraction_Pct'].sum()
    return min(risk_sum, 100.0)

lifestyle_factors = risk_factors[risk_factors['Category'] == 'Lifestyle']['Risk_Factor'].tolist()
potential_impact = oncological_risk_engine(lifestyle_factors)

print(f"ELITE INSIGHT: Combined modifiable lifestyle factors account for approximately {potential_impact:.1f}% of the population-level breast cancer burden.")

plt.figure(figsize=(12, 6))
sns.barplot(data=risk_factors[risk_factors['Category'] == 'Lifestyle'].sort_values('Population_Attributable_Fraction_Pct'), 
            x='Population_Attributable_Fraction_Pct', y='Risk_Factor', palette='crest')
plt.title("Strategic Intervention ROI: Modifiable PAF Ranking", fontweight='bold')
plt.show()
```

## 7. Supplemental Analytical Modules [NEW FINDINGS]
### 7.1 Multivariate Correlation Matrix
A high-fidelity evaluation of the interdependencies between clinical performance, screening coverage, and survival outcomes. This matrix identifies the strongest predictors of healthcare system success.

```python
plt.figure(figsize=(12, 10))
numeric_cols = ['Population_M', 'New_Cases_2022', 'Deaths_2022', 'Incidence_Rate_Per_100K', 
                'Mortality_Rate_Per_100K', 'Five_Year_Survival_Pct', 'Stage_I_II_Pct', 
                'Mammography_Coverage_Pct', 'Treatment_Access_Score', 'MIR']
corr_matrix = country_data[numeric_cols].corr()

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt=".2f", cmap='coolwarm', 
            linewidths=0.5, cbar_kws={"shrink": .8})

plt.title("Multivariate Correlation Matrix: Oncology System Interdependencies", fontweight='bold')
plt.show()
```

### 7.2 The Incidence-to-Mortality Paradox
Identifying global leaders that defy high-incidence trends through elite mortality capture. These countries represent the gold standard for oncological infrastructure.

```python
country_data['Paradox_Score'] = country_data['Incidence_Rate_Per_100K'] / (country_data['Mortality_Rate_Per_100K'] + 1)
paradox_leaders = country_data.sort_values('Paradox_Score', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(data=paradox_leaders, x='Paradox_Score', y='Country', palette="magma")
plt.title("Oncology Efficiency Leaders: Incidence-to-Mortality Paradox (Paradox Score)", fontweight='bold')
plt.xlabel("Paradox Score (High Incidence Efficiency)")
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.show()
```

### 7.3 Mammography Exposure vs. Five-Year Survival (Regression Analysis)
Quantifying the predictive power of screening coverage on clinical outcomes using linear modeling. This confirms the mathematical 'Dividend' of diagnostic exposure.

```python
from scipy import stats

x = country_data['Mammography_Coverage_Pct']
y = country_data['Five_Year_Survival_Pct']
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

plt.figure(figsize=(12, 8))
sns.regplot(x=x, y=y, scatter_kws={'alpha':0.5, 'color':VIBRANT_CYAN}, line_kws={'color':VIBRANT_PINK})
plt.annotate(f"R-squared: {r_value**2:.3f}\np-value: {p_value:.4f}", 
             xy=(0.05, 0.95), xycoords='axes fraction', 
             bbox=dict(boxstyle="round", fc=DARK_BG, ec=VIBRANT_CYAN, alpha=0.8), 
             color=TEXT_WHITE, fontweight='bold')

plt.title("Predictive Power of Mammography on 5-Year Survival", fontweight='bold', pad=30)
plt.xlabel("Mammography Coverage (%)", labelpad=20)
plt.ylabel("Five-Year Survival (%)", labelpad=20)
plt.subplots_adjust(top=0.88, bottom=0.15)
plt.show()
```

### 7.4 ROI Heatmap: Modifiable Lifestyle Factors vs. Unchangeable Risks
A side-by-side comparison of the Population Attributable Fraction (PAF). This heatmap prioritizes policy-driven lifestyle interventions against biological constants.

```python
risk_pivot = risk_factors.pivot_table(index='Risk_Factor', columns='Category', values='Population_Attributable_Fraction_Pct')

plt.figure(figsize=(10, 8))
sns.heatmap(risk_pivot, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Population Attributable Fraction (%)'})
plt.title("Strategic ROI Heatmap: Risk Modifiability Categories", fontweight='bold')
plt.xlabel("Risk Category")
plt.ylabel("Specific Risk factor")
plt.show()
```
